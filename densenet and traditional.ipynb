```
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('testX.csv')
```

```
# lets convert train and test Id as indices
train_data = train_data.set_index('Id')
test_data = test_data.set_index('Id')
```

```
# converting train data to numpy arrays
train_data = np.array(train_data, dtype = 'float32')
test_data = np.array(test_data,dtype = 'float32' )
```

```
# removing Label form train data
x_train = train_data[:,1:]/255
y_train = train_data[:,0]
x_test= test_data[:,:]/255
```

```
from sklearn.model_selection import train_test_split
x_train,x_validate,y_train,y_validate = train_test_split(x_train,y_train,test_size = 0.2,random_state = 12345)
np.shape(x_train)
np.shape(x_validate)
np.shape(y_train)
np.shape(y_validate)
```

```
np.shape(x_test)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten,BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import L1L2
#from keras.engine import Input, Model
```

```
dense_model = Sequential([
    Dense(1024,activation='relu'),
    Dense(512,activation='relu'),
    Dropout(0.4),
    Dense(128,activation='relu'),
    Dropout(0.25),
    Dense(32, activation="relu"),
    Dense(5,activation = 'softmax')
])
```

```
dense_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy'])
```

```
history = dense_model.fit(
    x_train,
    y_train,
    batch_size=1024,
    epochs=400,
    verbose=1,
    validation_data=(x_validate,y_validate),
)
dense_model.summary()
```

```
#defining image shape

image_rows = 28

image_cols = 28

batch_size = 1024
image_shape = (image_rows,image_cols,1)
```

```
# reshaping train and test and validation sets
x_train = x_train.reshape(x_train.shape[0],*image_shape)
x_test = x_test.reshape(x_test.shape[0],*image_shape)
x_validate = x_validate.reshape(x_validate.shape[0],*image_shape)
```

```
np.shape(x_train)
np.shape(x_validate)
np.shape(x_test)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten,BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import L1L2
#from keras.engine import Input, Model
```

```
cnn_model = Sequential([
    Conv2D(filters=16,kernel_size=5,activation='relu',padding='same',input_shape = image_shape),
    MaxPooling2D(pool_size=2) ,# down sampling the output instead of 28*28 it is 14*14
    Conv2D(filters=32,kernel_size=3,activation='relu',input_shape = image_shape,padding='same'),
    MaxPooling2D(pool_size=2),
    Dropout(0.2),
    Flatten(), # 1flatten out the layers
    Dense(1024,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation = 'softmax')
])
```

```
cnn_model.summary()
```

```
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy'])
```

```
history = cnn_model.fit(
    x_train,
    y_train,
    batch_size=batch_size,
    epochs=40,
    verbose=1,
    validation_data=(x_validate,y_validate),
)
```

```
evaluation=cnn_model.evaluate(x_validate,y_validate)
```
